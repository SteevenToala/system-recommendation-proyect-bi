```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from scipy.stats import gaussian_kde
from pathlib import Path

sns.set_theme(style='darkgrid', font_scale=1.05)
BG = '#12121f'; PANEL = '#1c1c2e'
C  = ['#7c6af7','#f97316','#22d3ee','#a3e635','#f43f5e','#fb923c','#34d399','#818cf8']
plt.rcParams.update({
    'figure.facecolor': BG, 'axes.facecolor': PANEL,
    'axes.labelcolor': 'white', 'xtick.color': 'white',
    'ytick.color': 'white',  'text.color': 'white',
    'axes.titlecolor': 'white', 'grid.color': '#2e2e4e',
    'legend.facecolor': PANEL, 'figure.dpi': 100,
})
print('Librerias cargadas OK')
```

## 1. Carga y resumen del dataset

```python
CSV_PATH = Path(r'D:/Proyectos/BI/10clase/sistemas_de_recomendacion/src/modules/alquileres_etiquetados.csv')
df = pd.read_csv(CSV_PATH)
FEATURES = ['ingreso', 'veces_alquilada_pelicula', 'precio_promedio_genero']
LABELS   = ['ingreso_alto', 'pelicula_popular', 'genero_alto_valor']
ALIAS    = {'ingreso_alto':'Ingreso Alto',
            'pelicula_popular':'Pelicula Popular',
            'genero_alto_valor':'Genero Alto Valor'}
print(f'{df.shape[0]:,} filas x {df.shape[1]} columnas')
df.head(10)
```

```python
df[FEATURES + LABELS].describe().round(3)
```

## 2. Distribucion de Features (Histograma + KDE)

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Distribucion de Features Numericas', fontsize=14, fontweight='bold')
for ax, col, color in zip(axes, FEATURES, C):
    data = df[col].dropna()
    ax.hist(data, bins=30, color=color, alpha=0.55, edgecolor=BG, density=True)
    kde = gaussian_kde(data)
    x = np.linspace(data.min(), data.max(), 300)
    ax.plot(x, kde(x), color='white', linewidth=2, label='KDE')
    ax.axvline(data.mean(),   color=C[1], lw=1.8, ls='--', label=f'Media {data.mean():.2f}')
    ax.axvline(data.median(), color=C[3], lw=1.8, ls=':',  label=f'Mediana {data.median():.2f}')
    ax.set_title(col, fontweight='bold')
    ax.set_xlabel('Valor')
    ax.set_ylabel('Densidad')
    ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

## 3. Funcion de Distribucion Acumulada (ECDF)

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('ECDF por Feature', fontsize=14, fontweight='bold')
for ax, col, color in zip(axes, FEATURES, C):
    data = np.sort(df[col].dropna().values)
    y    = np.arange(1, len(data) + 1) / len(data)
    ax.plot(data, y, color=color, linewidth=2.2)
    ax.fill_between(data, y, alpha=0.12, color=color)
    ax.axhline(0.25, color='white', lw=1,   ls='--', alpha=0.5, label='Q1  25%')
    ax.axhline(0.50, color=C[1],   lw=1.5, ls='--',             label='Mediana 50%')
    ax.axhline(0.75, color=C[3],   lw=1,   ls='--', alpha=0.5, label='Q3  75%')
    ax.set_title(f'ECDF — {col}', fontweight='bold')
    ax.set_xlabel('Valor'); ax.set_ylabel('Prob. acumulada')
    ax.set_ylim(0, 1); ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

## 4. Balance de Clases

```python
fig, axes = plt.subplots(1, 3, figsize=(13, 5))
fig.suptitle('Balance de Clases — Etiquetas Target', fontsize=14, fontweight='bold')
for ax, label in zip(axes, LABELS):
    counts = df[label].value_counts().sort_index()
    pcts   = counts / counts.sum() * 100
    bars   = ax.bar(['Clase 0', 'Clase 1'], counts.values,
                    color=[C[4], C[0]], edgecolor=BG, width=0.5)
    for bar, pct in zip(bars, pcts):
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 50,
                f'{pct:.1f}%', ha='center', fontsize=12,
                fontweight='bold', color='white')
    ratio = counts[0] / counts[1]
    ax.set_title(f'{ALIAS[label]}\nratio desbalance {ratio:.1f}:1',
                 fontsize=10, fontweight='bold')
    ax.set_ylabel('Cantidad')
    ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda v,_: f'{int(v):,}'))
plt.tight_layout()
plt.show()
```

## 5. Boxplot: Separacion entre Clases

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Boxplot: Feature por Clase (3 labels x 3 features)', fontsize=14, fontweight='bold')
for row, label in enumerate(LABELS):
    for col, feature in enumerate(FEATURES):
        ax = axes[row][col]
        g0 = df[df[label] == 0][feature].values
        g1 = df[df[label] == 1][feature].values
        bp = ax.boxplot([g0, g1], patch_artist=True,
                        medianprops=dict(color='white', linewidth=2.2),
                        flierprops=dict(marker='.', color='#ffffff44', markersize=2))
        bp['boxes'][0].set_facecolor(C[4]); bp['boxes'][0].set_alpha(0.75)
        bp['boxes'][1].set_facecolor(C[0]); bp['boxes'][1].set_alpha(0.75)
        ax.set_xticklabels(['Clase 0', 'Clase 1'])
        ax.set_title(f'{ALIAS[label]}\nvs {feature}', fontsize=8)
        if col == 0: ax.set_ylabel(feature)
plt.tight_layout()
plt.show()
```

## 6. Violin Plot: Densidad por Clase

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Violin Plot: Densidad por Clase', fontsize=14, fontweight='bold')
for row, label in enumerate(LABELS):
    for col, feature in enumerate(FEATURES):
        ax = axes[row][col]
        g0 = df[df[label] == 0][feature].dropna().values
        g1 = df[df[label] == 1][feature].dropna().values
        vp = ax.violinplot([g0, g1], positions=[1, 2], showmedians=True)
        vp['bodies'][0].set_facecolor(C[4]); vp['bodies'][0].set_alpha(0.65)
        vp['bodies'][1].set_facecolor(C[0]); vp['bodies'][1].set_alpha(0.65)
        vp['cmedians'].set_color('white'); vp['cmedians'].set_linewidth(2)
        ax.set_xticks([1, 2]); ax.set_xticklabels(['Clase 0', 'Clase 1'])
        ax.set_title(f'{ALIAS[label]}\nvs {feature}', fontsize=8)
        if col == 0: ax.set_ylabel(feature)
plt.tight_layout()
plt.show()
```

## 7. Histogramas Superpuestos (Clase 0 vs Clase 1)

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Histogramas Superpuestos por Clase', fontsize=14, fontweight='bold')
for row, label in enumerate(LABELS):
    for col, feature in enumerate(FEATURES):
        ax = axes[row][col]
        for val, color, lbl in [(0, C[4], 'Clase 0'), (1, C[0], 'Clase 1')]:
            data = df[df[label] == val][feature].dropna()
            ax.hist(data, bins=25, alpha=0.55, color=color,
                    edgecolor=BG, density=True, label=lbl)
        ax.set_title(f'{ALIAS[label]}\n{feature}', fontsize=8)
        ax.set_xlabel(feature); ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```

## 8. Mapa de Correlacion (Pearson y Spearman)

```python
cols_corr = FEATURES + LABELS
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Correlacion Pearson vs Spearman', fontsize=14, fontweight='bold')
for ax, method in zip(axes, ['pearson', 'spearman']):
    corr = df[cols_corr].corr(method=method)
    mask = np.triu(np.ones_like(corr, dtype=bool))
    sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
                linewidths=0.5, linecolor=BG, annot_kws={'size': 9},
                cbar_kws={'shrink': 0.75}, ax=ax)
    ax.set_title(f'Correlacion {method.capitalize()}', fontweight='bold')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=35, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```

## 9. Deteccion de Outliers (regla IQR)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Deteccion de Outliers — Regla IQR', fontsize=14, fontweight='bold')
for ax, feature, color in zip(axes, FEATURES, C):
    data = df[feature].dropna()
    Q1, Q3 = data.quantile(0.25), data.quantile(0.75)
    IQR = Q3 - Q1
    low, high = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    outliers = data[(data < low) | (data > high)]
    normal   = data[(data >= low) & (data <= high)]
    pct = len(outliers) / len(data) * 100
    ax.scatter(range(len(normal)),   normal.values,
               s=4, alpha=0.25, color=color, label='Normal')
    ax.scatter(range(len(outliers)), outliers.values,
               s=20, alpha=0.8, color=C[4], marker='x',
               label=f'Outlier {pct:.1f}%')
    ax.axhline(high, color=C[1], ls='--', lw=1.5, label=f'Lim sup {high:.2f}')
    ax.axhline(low,  color=C[3], ls='--', lw=1.5, label=f'Lim inf {low:.2f}')
    ax.set_title(feature, fontweight='bold')
    ax.legend(fontsize=7.5)
plt.tight_layout()
plt.show()
```

## 10. Pairplot (relacion entre features por etiqueta)

```python
# Ejecuta 3 pairplots (uno por etiqueta) — puede tardar ~30 seg
for label in LABELS:
    subset = df[FEATURES + [label]].copy()
    subset[label] = subset[label].map({0: 'Clase 0', 1: 'Clase 1'})
    g = sns.pairplot(subset, hue=label, diag_kind='kde',
                     plot_kws=dict(alpha=0.3, s=10, edgecolor='none'),
                     palette={'Clase 0': C[4], 'Clase 1': C[0]})
    g.figure.suptitle(f'Pairplot — {ALIAS[label]}',
                      y=1.01, color='white', fontsize=12)
    plt.show()
```

## 11. Barras 100%: Distribucion de Categoria por Etiqueta

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 6))
fig.suptitle('% de Clase Positiva por Categoria', fontsize=14, fontweight='bold')
for ax, label in zip(axes, LABELS):
    tabla = pd.crosstab(df['categoria_nombre'], df[label], normalize='index') * 100
    tabla.columns = ['Clase 0 %', 'Clase 1 %']
    tabla = tabla.sort_values('Clase 1 %', ascending=True)
    tabla.plot(kind='barh', stacked=True, ax=ax,
               color=[C[4], C[0]], edgecolor=BG, width=0.7)
    ax.set_title(ALIAS[label], fontweight='bold', fontsize=10)
    ax.set_xlabel('% registros'); ax.set_xlim(0, 100)
    ax.legend(fontsize=8, loc='lower right')
    ax.tick_params(axis='y', labelsize=8)
plt.tight_layout()
plt.show()
```

## 12. Estadisticas Descriptivas por Clase

```python
for label in LABELS:
    print(f'\n=== {ALIAS[label]} ===')
    display(df.groupby(label)[FEATURES].agg(['mean','std','min','max']).round(3))
```

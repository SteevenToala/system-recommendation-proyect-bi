# Clasificador Bayesiano (4 filas)
Usando `pelicula_popular` y `genero_alto_valor` para predecir `ingreso_alto`.

```
import pandas as pd
df = pd.read_csv('../src/modules/alquileres_etiquetados.csv')
f1, f2, target = 'pelicula_popular', 'genero_alto_valor', 'ingreso_alto'
total = len(df)
results = []
for x1 in [0, 1]:
    for x2 in [0, 1]:
        sub = df[(df[f1]==x1) & (df[f2]==x2)]
        c = len(sub)
        if c > 0:
            y1 = len(sub[sub[target]==1]); y0 = c - y1
            p_y1_x = y1/c; p_y0_x = y0/c; c_opt = 1 if p_y1_x > p_y0_x else 0
            err = (p_y0_x if c_opt==1 else p_y1_x) * (c/total)
            results.append({'x1':x1, 'x2':x2, 'FreqTotal':c, 'P(y=1|X)':p_y1_x, 'ClassOpt':c_opt, 'Error':err})
pd.DataFrame(results)
```

# Clasificador Bayesiano - Optimizacion $\psi_{opt}$ y Error $E_{out}$
Implementacion de las formulas de clase.

```
import pandas as pd
df = pd.read_csv('../src/modules/alquileres_etiquetados.csv')
f1, f2, target = 'pelicula_popular', 'genero_alto_valor', 'ingreso_alto'
res = []
for x1 in [0,1]:
    for x2 in [0,1]:
        sub = df[(df[f1]==x1) & (df[f2]==x2)]
        c = len(sub)
        if c > 0:
            y1 = len(sub[sub[target]==1]); y0 = c - y1
            p_x_y1 = y1/len(df); p_x_y0 = y0/len(df)
            psi_opt = 1 if p_x_y1 >= p_x_y0 else 0
            err = p_x_y0 if psi_opt==1 else p_x_y1
            res.append({'x1':x1, 'x2':x2, 'P(X,y=0)':p_x_y0, 'P(X,y=1)':p_x_y1, 'Psi_Opt':psi_opt, 'Error':err})
pd.DataFrame(res)
```
